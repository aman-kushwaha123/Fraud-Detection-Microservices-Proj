```python
from fastapi import FastAPI
import pickle
import numpy as np
```

```python
app=FastAPI()

model=pickle.load(open("fraud_model.pkl","rb"))

@app.post("/predict")
def predict(data:dict):
    features=np.array([[
        data["amount"],
        data["oldBalance"],
        data["userId"],
        data["deviceId"],
        data["time_stamp"],
        data["is_fraud"],
    ]])
    print("predicted")
    prediction=model.predict(features)[0]
    return {"isFraud":int(prediction)}
                      
        
```
